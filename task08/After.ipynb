```python
import sys
import cProfile
import numpy as np

class answer:
    def __init__(self, q):
        self.query = q
    def printing(self, x, y):
        x = x - 1
        res = self.query[x]
        s = 0
        for i in xrange(x, y):
            s = s + self.query[i]
            if res < s: res = s
            if s < 0: s = 0
        print(res)
    def changing(self, x, y):
        self.query[x - 1] = y
        
def func():
    f = open('test.txt')
    f.readline()
    p = answer((map(int, f.readline().split())))
    m = f.readline()
    i = 0
    while i < int(m):
        k = list(map(int, f.readline().split()))
        if int(k[0]) == 0:
            p.changing(k[1], k[2])
        else :
            p.printing(k[1], k[2])
        i = i + 1
    f.close()
cProfile.run('func()')
```
