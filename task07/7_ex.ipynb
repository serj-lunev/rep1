```python
from pandas import DataFrame, read_csv
import pandas as pd
import matplotlib.pyplot as plt
import sys

movies = pd.read_table('.movies.dat', header = None, delimiter = '::', names = ['Movie_id', 'Movie_name', 'Description'])
users = pd.read_table('.users.dat', header =None, delimiter = '::', names = ['User_id', 'Gender', 'Age', 'Occupation', 'ZipCode'])
ratings = pd.read_table('.ratings.dat', header = None, delimiter = '::', names = ['User_id', 'Movie_id', 'Rating', 'Timestamp'])
```

```python
'''1) 25 фильмов наиболее высоко оценены (сумма оценок по всем зрителям)'''

movie_name = movies[['Movie_id', 'Movie_name']]
mv_rating = ratings[['Movie_id', 'Rating']]
mv_rating = mv_rating.merge(movie_name, how = 'left', on = 'Movie_id')
mv_rating = mv_rating[['Movie_name', 'Rating']]
mv_rating_gr = mv_rating.groupby('Movie_name', as_index = False)
mv_rating = mv_rating_gr.sum()
sortedRt = mv_rating.sort('Rating', ascending=False)
del mv_rating, movie_name, mv_rating_gr
sortedRt.head(25)
```

```python
'''2) Фильмы получили самую высокую среднюю оценку?'''

mv_rating = ratings[['Movie_id', 'Rating']]
movie_name = movies[['Movie_id', 'Movie_name']]
mv_rating = mv_rating.merge(movie_name, how = 'left', on = 'Movie_id')
mv_rating = mv_rating[['Movie_name', 'Rating']]
mv_rating_gr = mv_rating.groupby('Movie_name', as_index = False)
mv_rating = mv_rating_gr.mean()
nameList = [mv_rating['Movie_name'][i] for i in range(len(mv_rating)) if mv_rating['Rating'][i] == mv_rating['Rating'].max()]
del mv_rating, movie_name, mv_rating_gr
DataFrame(nameList, columns = ['Movie'])
```

```python
'''3) О каких фильмах больше всего разногласий между мужчинами и женщинами?'''

mv_rating = ratings[['Movie_id', 'Rating', 'User_id']]
movie_name = movies[['Movie_id', 'Movie_name']]
user_sex = users[['User_id', 'Gender']]
mv_rating = mv_rating.merge(movie_name, how = 'left', on = 'Movie_id')
mv_rating = mv_rating[['Movie_name', 'Rating', 'User_id']]
mv_rating = mv_rating.merge(user_sex, how = 'left', on = 'User_id')
mv_rating = mv_rating[['Movie_name', 'Rating', 'Gender']]
mv_rating = mv_rating.groupby(['Movie_name', 'Gender'], as_index = False)
mv_rating = mv_rating.mean()
mv_rating = mv_rating.groupby('Movie_name', as_index = False)
mv_rating = mv_rating.std()
nameList = [mv_rating['Movie_name'][i] for i in range(len(mv_rating)) if mv_rating['Rating'][i] == mv_rating['Rating'].max()]
del mv_rating, movie_name, user_sex
DataFrame(nameList, columns = ['Movie'])
```

```python
'''4) Какие фильмы наиболее противоречивы среди разных возрастов?"'''

movie_name = movies[['Movie_id', 'Movie_name']]
mv_rating = ratings[['Movie_id', 'Rating', 'User_id']]
user_age = users[['User_id', 'Age']]
mv_rating = mv_rating.merge(movie_name, how = 'left', on = 'Movie_id')
mv_rating = mv_rating[['Movie_name', 'Rating', 'User_id']]
mv_rating = mv_rating.merge(user_age, how = 'left', on = 'User_id')
mv_rating = mv_rating[['Movie_name', 'Rating', 'Age']]
mv_rating = mv_rating.groupby(['Movie_name', 'Age'], as_index = False)
mv_rating = mv_rating.mean()
mv_rating = mv_rating.groupby('Movie_name', as_index = False)
mv_rating = mv_rating.std()
nameList = [mv_rating['Movie_name'][i] for i in range(len(mv_rating)) if mv_rating['Rating'][i] == mv_rating['Rating'].max()]
'''del rt1, mv1, us1'''
DataFrame(nameList, columns = ['Movie'])
```
